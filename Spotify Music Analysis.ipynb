```python
columns_corr = ['energy', 'tempo', 'danceability', 'loudness', 'valence','time_signature', 'speechiness', 'track_popularity', 
           'key', 'duration_ms', 'acousticness']
correlation = music_data[columns_corr].corr()
```

```python
genre_popularity_order = music_data.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='playlist_genre',y='track_popularity',data=music_data, order=genre_popularity_order, showmeans=True)
plt.title('Variation of Track Popularity by Genre')
plt.xlabel('Music Genre')
plt.ylabel('Track Popularity')
plt.show()
```

```python
genre_energy_order = music_data.groupby('playlist_genre')['energy'].mean().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='playlist_genre',y='energy',data= music_data, order=genre_energy_order, showmeans=True)
plt.title('Variation of Track Energy by Genre')
plt.xlabel('Music Genre')
plt.ylabel('Track Energy')
plt.show()
```

```python
genre_danceability_order = music_data.groupby('playlist_genre')['danceability'].mean().sort_values(ascending=False).index
plt.figure(figsize=(20,10))
sns.boxplot(x='playlist_genre',y='danceability',data=music_data, order=genre_danceability_order, showmeans=True)
plt.title('Variation of Track Danceability by Genre')
plt.xlabel('Music Genre')
plt.ylabel('Track Danceability')
plt.show()
```

```python
genre_loudness_order = music_data.groupby('playlist_genre')['loudness'].mean().sort_values(ascending=False).index
plt.figure(figsize=(20,10))
sns.boxplot(x='playlist_genre',y='loudness',data=music_data, order=genre_loudness_order, showmeans=True)
plt.title('Variation of Track Loudness by Genre')
plt.xlabel('Music Genre')
plt.ylabel('Track Loudness')
plt.show()
```

```python
genre_duration_order = music_data.groupby('playlist_genre')['duration_ms'].mean().sort_values(ascending=False).index
plt.figure(figsize=(20,10))
sns.boxplot(x='playlist_genre',y='duration_ms',data=music_data, order=genre_duration_order, showmeans=True)
plt.title('Variation of Track Duration by Genre')
plt.xlabel('Music Genre')
plt.ylabel('Track Duration')
plt.show()
```

```python
subgenre_popularity_order = music_data.groupby('playlist_subgenre')['track_popularity'].mean().sort_values(ascending=False).index
plt.figure(figsize=(20,10))
sns.boxplot(x='playlist_subgenre',y='track_popularity',data=music_data, order=subgenre_popularity_order, showmeans=True)
plt.title('Variation of Track Duration by SubGenre')
plt.xlabel('Music SubGenre')
plt.ylabel('Track Popularity')
plt.xticks(rotation=90)
plt.show()
```
